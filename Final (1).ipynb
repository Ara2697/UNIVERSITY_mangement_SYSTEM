```python
#some sort of License go here
# BSD 3-Clause License
# Copyright (c) 2023, Iffat Ara 
# All rights reserved.



#it starts right here
#------------------------------------------------------------------------------------------#
# Creating table
def create_table():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password="",database="student_info")
    cur=con.cursor()
    cur.execute('''
        CREATE TABLE IF NOT EXISTS basic_info (
            Roll_no varchar(225) PRIMARY KEY,
            Name varchar(225),
            Department varchar(225),
            Faculty varchar(225),
            Contact varchar(225),
            Address varchar(225),
            Gender varchar(225)
        )
    ''')
create_table()
#defining functions for easy call
def Labelfun(window,text="Youforgot",font=("Times New Roman",20,"bold"),bd=12,relief=tk.GROOVE,bg="lightgrey"):
    label=tk.Label(window,text=text,font=font,bd=bd,relief=relief,bg=bg)
    return label
def LabelFramefun(window,text="Youforgot",font=("Times New Roman",20,"bold"),bd=12,relief=tk.GROOVE,bg="lightgrey"):
    label=tk.LabelFrame(window,text=text,font=font,bd=bd,relief=relief,bg=bg)
    return label
def Framefun(window,bd=12,relief=tk.GROOVE,bg="lightgrey"):
    frame=tk.Frame(window,bd=bd,relief=relief,bg=bg)
    return frame

def Entryfun(frame1,font=("Arial"),bd=12,bg="white",textvariable=None):
    entry_frame=tk.Entry(frame1,font=("Arial",12),bd=12,bg="white",textvariable=textvariable)
    return entry_frame
def Entryfun(frame1,font=("Arial"),bd=12,bg="white",textvariable=None):
    entry_frame=tk.Entry(frame1,font=("Arial",12),bd=12,bg="white",textvariable=textvariable)
    return entry_frame
def Comboboxfun(lists,frame1,font=("Arial"),state="readonly",textvariable=None):
    entry_box=ttk.Combobox(frame1,font=font,state=state,textvariable=textvariable)
    entry_box["values"]=lists
    return  entry_box
def Buttonfun(button_frame,text="Forgot",bd=10,relief=tk.GROOVE,bg="lightgrey",width=20,command=None):
    button1=tk.Button(button_frame,text=text,bd=bd,relief=relief,bg=bg,width=width,command=command)
    return button1


import tkinter as tk
from tkinter import ttk
import pymysql
window=tk.Tk()
window.geometry("1200x600+0+0")
window.title("me") # default takes  (import tkinter as) "tk"
#windows text front-type size bold bd size bd shape  bd
# label=tk.Label(window,text="lets gooo",font=("Times New Roman",20,"bold"),bd=12,relief=tk.GROOVE,bg="grey")
label=Labelfun(window,"University System",("Times New Roman",20,"bold"),12,tk.SUNKEN,"lightgray")
label.pack(side=tk.TOP,fill=tk.X) # fill in x axis

#--------------------------------------------------------------------------------------------------#
#--------------------------------------------------------------------------------------------------#
#frist frame
#entry widget
frame=[] #frame[0],frame[1]
frame_name=["Input Form","Output"]
for i in range(2):
        frame.append(LabelFramefun(window,text=frame_name[i],font="Arial"))
        if(i==0):
            frame[i].place(x=20,y=90,width=400,height=500) 
        else:
            frame[i].place(x=20+400+40, y=90,width=700,height=500)

###---------------------###
# Assign variable
iterate=8-1
entery_label=["Roll_no","Name","Department","Faculty","contact","Adress","Gender"] #,"Fathers Name"
gender_list=["male","female","other"]
variable=[]
for i in range(iterate+2):
    variable.append(tk.StringVar())

# variable.append(tk.StringVar)
# variable.append(tk.StringVar)

#variable[0],variable[1],variable[2],variable[3],variable[4],variable[5],variable[6],variable[7]
#roll       ,Name       ,Department ,Faculty"   ,contact    ,Adress     ,Gender     2nd search


###---------------------###
# Input wigets
f1_label=[] #f1_label[0],f1_label[1].....f1_label[6]
f1_label_in=[]#f1_label_in[0],f1_label_in[1].....f1_label_in[6]

for i in range(iterate):
        f1_label.append(Labelfun(frame[0],text=entery_label[i],font="Arial",bd=12,bg="lightblue",relief=tk.FLAT))
        f1_label[i].grid(row=i,column=0,padx=2,pady=2,sticky=("w"))
        if(entery_label[i]=="Gender"):
            
            f1_label_in.append(Comboboxfun(gender_list,frame[0],font=("Arial"),state="readonly",textvariable=variable[i]))
            f1_label_in[i].grid(row=i,column=1,padx=2,pady=2)
            continue
        f1_label_in.append(Entryfun(frame[0],textvariable=variable[i]))
        f1_label_in[i].grid(row=i,column=1,padx=2,pady=2,sticky=("w"))

        
###---------------------###
# Button Funtion & DB connector


def add_btn():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password='',database=database)
    sql = "INSERT INTO basic_info VALUES (%s,%s,%s,%s,%s,%s,%s)"
#    val = (variable[0].get(),variable[1].get(),variable[2].get(),variable[3].get(),variable[4].get(),variable[5].get(),variable[6].get())
#     new_list = variable[:]
#     existing_list.pop()    
    new_val=()
    for i in range(7):
        new_val= new_val + (variable[i].get(),)   
    cur = con.cursor()
    cur.execute(sql, new_val)
    con.commit()
    con.close()

def show_all():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password="",database="student_info")
    cur=con.cursor()
    cur.execute("select * from basic_info")
    rows=cur.fetchall()
    if len(rows)!=0:
        student_table.delete(*student_table.get_children())
        for row in rows:
            student_table.insert('',tk.END,values=row)
        con.commit()
    con.close()

def clear():
    for i in range(7):
        variable[i].set("")
def update():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password="",database="student_info")
    cur=con.cursor()
    sql="update basic_info set Name=%s,Department=%s,Faculty=%s,Contact=%s,Address=%s,Gender=%s where Roll_no=%s"
    new_val=()
    
    for i in range(1,7):
        new_val= new_val + (variable[i].get(),) 
    new_val= new_val + (variable[0].get(),) 
    print(sql,new_val)
    cur.execute(sql,new_val)
    con.commit()
    con.close()

def delete():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password="",database=database)
    cur=con.cursor()
    cur.execute("delete from basic_info where roll_no=%s",variable[0].get())
    con.commit()
    con.close()
def search():
    database="student_info"
    table="basic_info"
    con=pymysql.connect(host="localhost",user="root",password="",database="student_info")
    cur=con.cursor()
    cur.execute("select * from basic_info where "+str(variable[len(variable)-2].get())+" like '%"+str(variable[len(variable)-1].get())+"%'")
    rows=cur.fetchall()
    student_table.delete(*student_table.get_children())
    for row in rows:
        student_table.insert('',tk.END,values=row)
    con.commit()
    con.close()

##----------------------------------------------------------##
# button widgets

button_frame=[]
No_of_button=4+1
button1=[]
button_text=["Add","Delete","Modify","Clear"]
# func_list=["add_btn()","del_btn()","mod_btn()","clr_btn()"]
func_list=["add_btn","None","None","None"]
# button_frame=tk.LabelFrame(window,bd =12,relief=tk.GROOVE,bg="lightgrey")
# button_frame.place(x=35,y=450,width=360,height=120)
for i in range(1):
        button_frame.append(LabelFramefun(frame[0],text=None,bg="red"))
        button_frame[i].place(x=5,y=332,width=360,height=120)
        
add = ttk.Button(button_frame[0],text="Add",width=10,command=add_btn)
add.grid(row=0,column=0,padx=5,pady=10)

update = ttk.Button(button_frame[0],text="Update",width=10,command=update)
update.grid(row=0,column=1,padx=8,pady=10)

delete = ttk.Button(button_frame[0],text="Delete",width=10,command=delete)
delete.grid(row=0,column=2,padx=8,pady=10)

clear= ttk.Button(button_frame[0],text="Clear",width=10,command=clear)
clear.grid(row=0,column=3,padx=8,pady=10)






#---------------------------------------------------------------------------#
#second frame
#search widget


button2=[]
for i in range(1):
        
        button_frame.append(Framefun(frame[1]))
        button_frame[i+1].pack(side=tk.TOP,fill=tk.X)
        for j in range(No_of_button): 
            if(j==0):           
                button2.append(Labelfun(button_frame[i+1],text="Search",font=("arial",12),relief=None)) #default font is samller than 12 without Tk relief it will seem the lable took all x
                button2[j].grid(row=0,column=j,padx=2,pady=2,sticky=("w"))
                continue
            elif(j==1):           
                button2.append(Comboboxfun(entery_label,button_frame[i+1],textvariable=variable[len(variable)-2])) #default font is samller than 12 without Tk relief it will seem the lable took all x
                button2[j].grid(row=0,column=j,padx=7,pady=2,sticky=("w"))                
                continue
            elif(j==2):
                #Button2.append(Entryfun(button_frame[i+1],textvariable=variable[len(variable)-1]))
                button2.append(tk.Entry(button_frame[i+1],font=("Arial",12),bg="white",textvariable=variable[len(variable)-1],width=15))
                button2[j].grid(row=0,column=j,padx=7,pady=2) 
            elif(j==3):
                Status = "Go" 
                button2.append(Buttonfun(button_frame[i+1],text=Status,width=8,command=search))
                button2[j].grid(row=0,column=j,padx=7,pady=2)
            else :
                Status = "Search All"
                button2.append(Buttonfun(button_frame[i+1],text=Status,width=8,command=show_all))
                button2[j].grid(row=0,column=j,padx=7,pady=2)

                
#---------------------------------------------------------------------------#
#second frame
#table widget
columns=tuple(entery_label)
result_frame=Framefun(frame[1])
result_frame.pack(fill=tk.BOTH,expand=True)

y_scroll=tk.Scrollbar(result_frame,orient=tk.VERTICAL)
x_scroll=tk.Scrollbar(result_frame,orient=tk.HORIZONTAL)
student_table=ttk.Treeview(result_frame,columns=tuple(entery_label),show='headings',yscrollcommand=y_scroll.set,xscrollcommand=x_scroll.set) #without show heading one coloumn at the start remains empty

y_scroll.config(command=student_table.yview)# connecting scroll with table view
x_scroll.config(command=student_table.xview)


y_scroll.pack(side=tk.RIGHT,fill=tk.Y)
x_scroll.pack(side=tk.BOTTOM,fill=tk.X)


for i in range(iterate):
    student_table.heading(columns[i],text=columns[i])

for i in range(iterate):
    student_table.column(columns[i],width=150)
#student_table['show']='headings'

student_table.pack(fill=tk.BOTH,expand=True)

# show_all()




window.mainloop()         
```
